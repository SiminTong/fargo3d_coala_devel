```python
dens = []
vr = []
stokes = []

n = 1
Ndust = 20

gasdens = fromfile("../../outputs/fargo_multifluid/gasdens{0:d}.dat".format(n))

for id in range(1,Ndust+1):
    dens_ = fromfile("../../outputs/fargo_multifluid/dust{0:d}dens{1:d}.dat".format(id,n))
    vr_ = fromfile("../../outputs/fargo_multifluid/dust{0:d}vy{1:d}.dat".format(id,n))
    cs_ = fromfile("../../outputs/fargo_multifluid/dust{0:d}energy{1:d}.dat".format(id,n))
    dens.append(dens_)
    vr.append(vr_)
    stokes.append(cs_/gasdens)

rmin = loadtxt("../../outputs/fargo_multifluid/domain_y.dat")[3:-3]
rmed = 0.5*(rmin[1:]+rmin[:-1])
```

```python
### plot radial velocity 
for m in range(Ndust):
    plot(rmed, vr[m])

show()
```

```python
### plot radial velocity 
for m in range(Ndust):
    plot(rmed, dens[m])

show()
```

```python
### plot radial velocity 
for m in range(Ndust):
    plot(rmed, stokes[m])
yscale('log')
show()
```

```python
### plot radial velocity 
for m in range(Ndust):
    plot(m, dens[m][256], 'o')
yscale('log')
show()
```
